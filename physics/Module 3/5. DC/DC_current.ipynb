# 1. Circuits

* **Circuit**: a interconnection of electric devices.
* **Electric component**: characterized by voltage-current relation among terminals, called **characteristic curve**.
    - pasive
    - active
* **Terminal**:endpoint of an electric component
    - two-terminal devices: resistance, capacitor, coil...
    - tree-terminal devices: transistors, rely, some switches..
* **Connection**: ideal wire with zero resistance.
* **Node/junction**: any point on a circuit where the terminals of two or more circuit elements meet.
* **Ground**: a reference node, at which the potential is 0.
* **Branch**: any path between two nodes without nodes in between.
* **Loop**: any closed path in a circuit.
* **Mesh**: a loop that does not enclose other loops.

Types of circuits
* __DC current__: direction of current remains constant.
* __AC current__: current oscillates. 

Types of circuits
* __linear__: all components have linear characteristic curve. Given a linear combination of input voltages $\mathrm{v}_{\mathrm{i}}=\mathrm{A} \mathrm{v}_{\mathrm{i} 1}+\mathrm{B} \mathrm{v}_{\mathrm{i} 2}$, the output is also a linear combination $\mathrm{v}_{o}=\mathrm{A} f\left(\mathrm{v}_{\mathrm{i} 1}\right)+\mathrm{B} f\left(\mathrm{v}_{\mathrm{i} 2}\right)$
* __non-linear__: some components are non-linear.

**Circuit analysis**: given a circuit, calculate voltage or currents at some node or branch.

**Circuit design**: given an input-output relation, combine electronic devices fulling such requirement.

## 2.2 Capacitor
Accumulates electric charge. Measured in farads (F). 

Symbol
<center><img src="symbolCap.png" width="200" /></center>

Curve I-V:

$$q(t)=CV(t)\implies I(t)=\frac{dq}{dt}=C \cdot \frac{d V(t)}{d t}$$

$$V(t)=\frac{1}{C} \cdot \int_{-\infty}^{t} I(\tau) \cdot d \tau=V(0)+\frac{1}{C} \cdot \int_{0}^{t} I(\tau) \cdot d \tau$$

Voltage depends on the sum (integral) of currents that went through it previously, this is, capacitors has memory of the past.


Capacitors in **series**. Since $V_{a b}=V=V_{1}+V_{2}=Q\left(\frac{1}{C_{1}}+\frac{1}{C_{2}}\right)$, then $\frac{1}{C_{\mathrm{eq}}}=\frac{1}{C_{1}}+\frac{1}{C_{2}}+\frac{1}{C_{3}}+\cdots=\sum_j \frac{1}{C_j}$

<center><img src="seriesCap.png" width="300" /></center>

Capacitors in **parallel**. Since $Q=Q_{1}+Q_{2}=\left(C_{1}+C_{2}\right) V$ then $C_{\mathrm{eq}}=C_{1}+C_{2}+C_{3}+\cdots=\sum_j C_j$

## 2.3 Inductor

Found in solenoids, but also en electric motors. Measured in henry (H). 

Symbol:

<center><img src="solenoid.png" width="200" /></center>

Curve I-V:

$V(t)=L \cdot \frac{dI(t)}{d t}$

$I(t)=I(0)+\frac{1}{L} \cdot \int_{0}^{t} V(\tau) \cdot d \tau$

Current depends on the sum (integral) of previous voltage, it has memory of the past.

Inductors in **series**. Since $V_{a b}=V=V_{1}+V_{2}=\left(L_1+L_2\right) \frac{dI(t)}{d t}$, then $L_{\mathrm{eq}}=L_{1}+L_{2}+L_{3}+\cdots=\sum_j L_j$

Capacitors in **parallel**. Since $I=I_{1}+I_{2}\implies=\frac{dI}{dt}=\frac{dI_1}{dt}+\frac{dI_2}{dt}=V_{ab}(\frac{1}{L_1}+\frac{1}{L_2})=V_{ab}\frac{1}{L}$, then ${1/L_{\mathrm{eq}}}=\frac{1}{L_{1}}+\frac{1}{L_{2}}+\frac{1}{L_{3}}+\cdots=\sum_j \frac{1}{L_j}$

<center><img src="inductors_series_paral.png" width="400" /></center>

Recap:
    
$$V=RI$$
$$V(t)=L \cdot \frac{dI(t)}{d t}$$
$$I(t)=C \cdot \frac{d V(t)}{d t}$$

# 5. Kirchhoff's rules

__Kirchhoff's junction rule (KCL)__: The sum of currents into any node is zero. This is due to the __conservation of charge__ (the total charge flowing into the junction per unit time must equal the total charge leaving per unit time). Sign convention: A current flowing in (out) is positive (negative)

$$\sum_j I_j=0$$

<center><img src="Kir1.png" width="500" /></center>

__Kirchhoff's loop rule (KVL)__: The algebraic sum of the potential differences in any loop is zero. This is due to the __conservation of energy__.

$$\sum_j V_j=0$$


## 5.1 How to use them?

1. Identify relevant nodes of the circuit and pick one as a reference (usually the ground GND, where V=0).
2. Assume a direction for the current in each branch.
3. Apply KCL to every node except reference. 

Example:


![](exampleKCL.png)
$\begin{equation}(1) I_{1}+I_{2}+I_{3} \\
(2) I_{2}=I_{5}+I_{4} \\
(3) I_{3}+I_{4}+I_{0}=0\end{equation}$

3 equations, 5 unknowns (currents).

4. Identify relevant loops in the circuit
5. Choose a direction for every loop (clockwise or counter-clockwise).
6. Starting at any point, go around a loop adding proper **voltage drops** (emfs, IR...), following a sign convention. Voltage is positive if we travel from + to -; negative otherwise. In resistors, capacitors and inductors, polarity is given by current directions (assigned previously).
 
<center><img src="signConvention.png" width="500" /></center>

Example:


![](exampleKVL.png)
\begin{array}{l}
(1321) \: I_{3} R_{3}-I_{4} R_{4}-V_{a}-I_{2} R_{2}=0 \\
(0120) \: -I_{1} R_{1}+I_{2} R_{2}+V_{a}+I_{5} R_{5}=0
\end{array}

+2 equations $\implies$ 5 equations and 5 unknowns. Notice we have not written any equation for loop (023); it is not necessary and we don't know the voltage drop at the current source.

The solution to the system of equations allow us to know the intensity current at any branch. To determine the potential at any node we travel from the node to the reference (along the simplest path).

$$V_{1}=I_{1} R_{1} ;\: V_{2}=I_{5} R_{5} ;\: V_{3}=-I_{3} R_{3}+I_{1} R_{1}$$

# 6. Transient and steady state

When we switch on a circuit, it goes through two distinct regimes: firstly a **transient state** and secondly a **steady state**. In the transient, currents and voltages vary with time. In the steady state, magnitudes are either constant or periodic in time (like in AC circuits that we shall study in the next unit).

In a steady state, the total charge in every segment of the circuit must be constant. Thus, the flow of charge coming out of a segment at any moment must be the same as the flow of charge going into the segment at the other end: the current is the same across all sections of the circuit. 

Previously we have analyzed circuits that are time-independent. This is the usual state of a circuit, since transient states are normally short-lived. However, it's important to understand the transient of a few important circuits.

## 6.1 RC circuit (charging)

|   |    |
|---|---|
|![](RC1.png)|![](RC2.png)|

Kirchhoffs loop rule: $$\mathcal{E}-i(t) R-\frac{q(t)}{C}=0$$

By definition: $i(t)=d q(t) / d t$.

$$\frac{d q}{d t}=\frac{\mathcal{E}}{R}-\frac{q}{R C}=-\frac{1}{R C}(q-C \mathcal{E})$$
$$\frac{d q}{q-C \mathcal{E}}=-\frac{d t}{R C}$$
$$\int_{0}^{q} \frac{d q^{\prime}}{q^{\prime}-C \mathcal{E}}=-\int_{0}^{t} \frac{d t^{\prime}}{R C}$$
$$\ln \left(\frac{q-C \mathcal{E}}{-C \mathcal{E}}\right)=-\frac{t}{R C}$$

|  charge in charging RC | current in charging RC |
|------------------------|------------------------|
|$$q(t)=C \mathcal{E}\left(1-e^{-t / R C}\right)=Q_{\mathrm{f}}\left(1-e^{-t / R C}\right)$$|$$i(t)=\frac{d q}{d t}=\frac{\mathcal{E}}{R} e^{-t R C}=I_{0} e^{-t / R C}$$|

We define the **time constant** or **relaxation time**

$$\tau=RC$$

<center><img src="RC3.png" width="300" /></center>
- Units of time (check it!)
- After a time equals to RC, the capacitor has reached $(1-1/e)\approx 63,2\%$ of its capacity.
- The larger the time constant, the more time it takes for the circuit to charge.

## 6.2 RC circuit (discharging)
After is the capacitor has been charged with a charge $Q_0$, we remove the battery.

Kirchhoffs loop rule: $$i(t) R-\frac{q(t)}{C}=0$$

$$i=\frac{d q}{d t}=-\frac{q}{R C}$$
$$\int_{Q_{0}}^{q} \frac{d q^{\prime}}{q^{\prime}}=-\frac{1}{R C} \int_{0}^{t} d t^{\prime}$$
$$\ln \frac{q}{Q_{0}}=-\frac{t}{R C}$$

|  charge in discharging RC | current in discharging RC |
|------------------------|------------------------|
|$$q(t)=Q_{0} e^{-t / (R \mathcal{C})}$$|$$i(t)=\frac{d q}{d t}=-\frac{Q_{0}}{(R C)} e^{-t / (R C)}=I_{0} e^{-t  /(R C)}$$|

# 7. Electric power

Electric power is work per unit time done by the electric field. In a circuit, this coincides with the rate at which energy is
either delivered to or extracted from a circuit element. 

$$P=\frac{dW}{dt}=\frac{VdQ}{dt}=VI$$

- Measured in Watts (W) in honor of James Watt. $(1 \mathrm{J} / \mathrm{C})(1 \mathrm{C} / \mathrm{s})=1 \mathrm{J} / \mathrm{s}=1 \mathrm{W}$

## 7.1 Power dissipated by a resistor

$$P=V_{a b} I=I^{2} R=\frac{V_{a b}^{2}}{R}$$

- The electric energy dissipated by resistor is transformed into other energy forms. In a pure resistance, temperature of the resistor increases and there is a flow of heat out of it. In a light bulb, part of the energy is transformed into light.
- Every resistor has a power rating, the maximum power the device can dissipate without becoming overheated and damaged.

## 7.2 Power absorbed/supplied by a capacitor

$$p(t)=v(t) \cdot i(t)=v(t) \cdot C \frac{d v(t)}{d t}=\frac{d}{d t}\left(\frac{1}{2} C v_{c}^{2}(t)\right)$$

- Power can be positive (absorption of energy) or negative (supply of energy) depending on the sign of the derivative. 

## 7.3 Power absorbed/supplied by an inductor

$$p(t)=i(t) \cdot v(t)=i(t) \cdot L \frac{d i(t)}{d t}=\frac{d}{d t}\left(\frac{1}{2} L i_{L}^{2}(t)\right)$$

- Power can be positive (absorption of energy) or negative (supply of energy) depending on the sign of the derivative. 

## 7.4 Power supplied by a source

A source described by an emf $\mathcal E$ and an internal resistance $r$. The rate at which the source delivers electrical energy to the remainder of the circuit is

$$P=V_{a b} I=\mathcal{E} I-I^{2} r$$

The first term is the work done by the battery, and the second term is the power dissipated by internal resistance of the battery.

# 8. Systematic circuit analysis

## 8.1 Mesh analysis

1. Identify meshes in the circuit and assign then a mesh current (unknowns).
2. Apply KVL as many times as unknowns are.
3. Once the system of eq.s is solved, the net current in each branch will be the sum of mesh currents flowing through that branch.

Example


![](mesh.png)

\begin{array}{l}
(mesh1)\: I_{1} R_{3}+\left(I_{1}-I_{3}\right) R_{4}-V_{a}+\left(I_{1}-I_{2}\right) R_{2}=0 \\
(mesh2)\: I_{2} R_{1}+\left(I_{2}-I_{1}\right) R_{2}+V_{a}+\left(I_{2}-I_{3}\right) R_{5}=0 \\
(mesh3)\: \left(I_{3}-I_{2}\right) R_{5}+\left(I_{3}-I_{1}\right) R_{4}+I_{3} R_{6}=0
\end{array} 

The current flowing through $R_2$ would be $I_1-I_2$

Remarks:

- The number of independent equations satisfy $# meshes= # branches-#nodes+1$.
- One can actually rely on loops instead of meshes when that is convenient.
- In the case of having a current source, one of the equations is trivial.


![](mesh2.png)
\begin{array}{l}
(mesh1)\: I_{1} R_{3}+\left(I_{1}-I_{3}\right) R_{4}-V_{a}+\left(I_{1}-I_{2}\right) R_{2}=0 \\
(mesh2)\: I_{2} R_{1}+\left(I_{2}-I_{1}\right) R_{2}+V_{a}+\left(I_{2}-I_{3}\right) R_{5}=0 \\
(mesh3)\: I_{3}=-I_{0}
\end{array}

## 9.2 Thevenin's theorem

Any linear circuit containing several voltages and resistances can be replaced by just one single voltage in series with a single resistance. 

|    |     |
|----|-----|
|<img src="thevenin2.png" width="300"/>|<img src="thevenin.png" width="300"/>|

- The equivalent voltage $V_{th}$ is the voltage obtained at terminals A-B of the network with terminals A-B open circuited.
- The equivalent resistance $R_{th}$ is the resistance that the circuit between terminals A and B would have if all ideal voltage sources in the circuit were replaced by a short circuit and all ideal current sources were replaced by an open circuit.


Thevenin theorem can be used to reduce a complex circuit into a simple circuit, which simplifies further analysis.
