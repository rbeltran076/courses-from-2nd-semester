### Problem 2

Two concentric circular loops of wire lie on a tabletop, one inside the other. The inner wire has a diameter of 20.0 cm and carries a clockwise current of 12.0 A, as viewed from above, and the outer wire has a diameter of 30.0 cm. What must be the magnitude and direction (as viewed from above) of the current in the outer wire so that the net magnetic field due to this combination of wires is zero at the common center of the wires?

*Answer*: $$
B_{1}=\frac{\mu_{0} I}{2 R_{1}}, B_{2}=\frac{\mu_{0} I}{2 R_{2}}
$$
$$
\frac{\mu_{0} I_{1}}{2 R_{1}}=\frac{\mu_{0} I_{2}}{2 R_{2}}
$$
$$
I_{2}=\left(\frac{R_{2}}{R_{1}}\right) I_{1}=\left(\frac{15.0 \mathrm{~cm}}{10.0 \mathrm{~cm}}\right)(12.0 \mathrm{~A})=18.0 \mathrm{~A}
$$
$I_2$ is counterclockwise.

### Problem 3
A solid conductor with radius $a$ is surrounded by a conducting tube with inner radius $b$ and out radius $c$ as shown in the figure. The central conductor and tube carry equal currents $I$ in opposite directions and uniformly distributed over their cross sections. Find the magnetic field in every point of space.
<center><img src="module1bp3.png" width="300"/></center>

*Answer*: 


For $r<a$:
$$B(2\pi r)=\mu_0(I_1)_{encl}=\mu_0\frac{I}{\pi a^2}(\pi r^2)\implies \vec B=\frac{\mu_{o} I}{2 \pi a^{2}} r\quad\hat u_\varphi$$
For $a<r<b$
$$B(2\pi r)=\mu_0I \implies \vec B=\frac{\mu_{o} I}{2 \pi r} u_\varphi$$
For $b<r<c$
$$B(2\pi r)= \mu_0(I_1-(I_2)_{encl})\implies B=\mu_0(I-\frac{I\pi(r^2-b^2)}{\pi(c^2-b^2)})\implies \vec B=\frac{\mu_{o} I}{2 \pi r}\left(\frac{c^{2}-r^{2}}{c^{2}-b^{2}}\right)\hat u_\varphi  $$
For $c<r$
$$B(2\pi r)= \mu_0(I_1-I_2)=0\implies B=0$$

### Problem 6

The figure shows a silver ribbon with $z_1=11.8mm$ and $y_1=0.23mm$ carrying a current of $120 A$ in the positive x-direction. The ribbon lies in a uniform magnetic field in the positive y-direction, with magnitude 0.95 T. 
<center><img src="hall_ribbon.png" width="250"/></center>

Knowing that for silver there are $5.85\times 10^{28}$ free electrons per cubic meter, find

* a) the drift velocity of the electrons in the x-direction. 
* b) the magnitude and direction of the electric field in the z-direction due to the Hall effect. 
* c) the Hall emf.

*Answer*

a) Using the definition of the current density

$$
J_{x}=\frac{I}{A}=\frac{I}{y_{1} z_{1}}=\frac{120 \mathrm{~A}}{\left(0.23 \times 10^{-3} \mathrm{~m}\right)(0.0118 \mathrm{~m})}=4.42 \times 10^{7} \mathrm{~A} / \mathrm{m}^{2}
$$

Additionally, we can relate the current density with the drift velocity of electrons.

$$
v_{\mathrm{d}}=\frac{J_{x}}{nq}=\frac{4.42 \times 10^{7} \mathrm{A} / \mathrm{m}^{2}}{\left(5.85 \times 10^{28} / \mathrm{m}^{3}\right)\left(-1.602 \times 10^{-19} \mathrm{C}\right)}=-4.7 \times 10^{-3} \mathrm{m} / \mathrm{s}=-4.7 \mathrm{~mm} / \mathrm{s}
$$

b) As for the module of the electric field, we know that the magnetic and electric forces must be equal in module with opposite directions (in equilibrium). Then,

$$
\begin{array}{l}
|q E_{z}|=|q v_{\mathrm{d}} B_{y}| \\
|E_{z}|=|v_{\mathrm{d}} B_{y}|=\left(4.7 \times 10^{-3} \mathrm{~m} / \mathrm{s}\right)(0.95 \mathrm{~T})=4.5 \times 10^{-3} \mathrm{~V} / \mathrm{m}
\end{array}
$$

Using the right-hand rule, the magnetic force is along the positive z-axis, which means that the electric force must be along the negative z-axis. As the charge carrier is negative (electrons), the electric field must be in the positive z-axis. All in all,

$$\vec E=+4.5 \times 10^{-3} \mathrm{~V} / \mathrm{m}\quad\hat z$$

c) The Hall emf is the potential difference between the two edges of the strip (from $z=0$ and $z=z_1$) that results from the electric field calculated in part b). $V_{hall} =Ez_1 =(4.5 \times 10^{-3} V/m)(0 0118 m)=53 \mu V$
