# True or false (and why)


1. Insulators can be charged by induction.
    * False. Only conductors
2. An electrically charged insulator and a neutral conductor attract each other.
    * True (by polarization of charges in the conductor)
3. The electric force exerted on $q_1$ by $q_2$ does not change even if another charge $q_3$ shows up.
    * True. Superposition principle.
4. The acceleration of a charge located at a certain point in space where there exists an electric field lies in the direction of the electric field line that passes through that point. 
    * True. Second Newton's law.
5. A body cannot create an electric field unless it has a net electric charge.
    * False. Think of a dipole.
6. A dipole whose moment is perpendicular to an external electric field does not move.
    * False. It rotates.
7. The existence of a nucleus contradicts the Coulomb force.
    * True. That's why we need nuclear forces.
8. If the electric flux through a closed surface is zero then there cannot be any charge inside the surface.
    * False. There can be charges as long as the net charge is zero.
9. If the electric flux through a closed surface is zero, the electric field must be zero at any point over that surface.
    * False.
10. If the electric flux through a closed surface is zero, then number of field lines going out of the surface must equal the number of lines going in.
    * True
11. Gauss law is only valid for symmetrical charge distributions.
    * False.
12. A gaussian surface is one such that the electric field is constant over it.
    * False
13. Close enough to a uniformly charged disk, the electric field resembles the field created by an infinitely charged sheet.
    * True
14. If the electric field is zero at any point of a closed surface, then there must be no net charge inside the surface.
    * True
15. The electric potential can be positive or negative depending on the sign of the source charge.
    * True, by definition
16. The net potential at any given point created by a collection of charges is the vectorial sum of the potential created by each charge independently.
    * False. Algebraic sum.
17. It's impossible to have a negative electric potential between the space of two positive charges.
    * False. Potential depends on the chosen origin of energies.
18. The potential energy of a collection of charges is $U=\frac{1}{4 \pi \epsilon_{0}} \sum_{i,j} \frac{q_{i} q_{j}}{r_{i j}}$.
    * False. it should be $\sum_{i<j}$ to avoid double counting pairs.
19. We put three identical charges in the vertices of a triangle. In the center of the triangle, both the electric field and the potential are zero. 
    * False. Electric field is zero, but not the potential.
20. In presence of an electric field, it's impossible to move a charge without doing work against the field.
    * False. You could move the charge perpendicular to the field.
21. Capacitor 1 has twice as much capacitance as capacitor 2 given the same charge. Therefore, capacitor 1 has also twice as much voltage as capacitor 2
    * False. Voltage in 2 must be half.
22. Capacitor 1 has twice as much charge as capacitor 2 given the same voltage. Consequently, the energy stored in capacitor 2 doubles the energy of capacitor 1.
    * False. Energy in 1 doubles 2
23. Capacitance decreases when a dielectric is put between the plates of a capacitor.
    * False. It increases
23. Given a capacitor connected to a battery, its charge decreases when a dielectric is put between its plates.
    * False. Dielectric gets polarized the capacitor admits more charge to keep the electric field between the plates.

# Questions

1. Suppose you have a positive charge $q$ that you want to distribute in two different bodies. How do you have to distribute the electrical charge between them so that electrical repulsion is maximum?
    * Electric force is proportional to the product of charges $F\propto (q-x)x\implies dF/dx\propto(q-2x)$. Then $dF/dx=0\implies x=q/2$. Equal distribution between bodies.
2. Choose the figure that best represents Coulomb force as a function of distance.
<center><img src="graphF.png" width="600" /></center>
    * A. $F\propto 1/r^2$.
3. The dielectric strength of air is $3 MV/m$, what should be the minimum separation between the plates of a capacitor to avoid dielectric breakdown given a voltage of $4kV$?
    * minimum separation translates into maximum electric field between the plates, which can be written in terms of voltage as $d_{min}=\Delta V/E_{max}=3\times 10^3(V)/4\times 10^6(V/m)=0.75mm$.
4. A charge $Q_1$ is located on the positive $x$ axis, and another charge $Q_2$ is located on the same axis but further away from the origin $O$. Knowing that the total electric charge is zero in $O$, what can we state about the signs of the charges? and what about their magnitudes?
    * For the net electric field to be zero at the origin, signs of charges must be opposite. Additionally, the smallest charge should be closer to the origin so $|Q_2|>|Q_1|$.
5. A balloon of radius $R$ if full of a gas with charge $Q$. The balloon is injected with another neutral gas so that the radius doubles. How does the electric field $E$ vary at a distance $R$ from the center of the balloon?
    * $\rho_1=\frac{Q}{4/3\pi R^3}$ and $\rho_2=\frac{Q}{4/3\pi (2R)^3}$. Then $\rho_1=8\rho_2$. Gauss law implies $E_1/E_2=\rho_1/\rho_2=8$
6. For parallel-plate capacitor with $d=1mm$, how much surface do we need to store $1F$?
    * $A=Cd/\epsilon_0=1F\times 10^{-3}m/8.85\times10^{-12}(F/m)=10km^2$. A square of side $10km$.
7. A little metallic sphere is placed between the plates of a capacitor. What is the value of the electric field and the potential inside the sphere?
    * Free charges inside the sphere rearange so that the electric field inside the sphere is zero, and the potential is constant.
8. Discuss what happens to a charged parallel-plate capacitor when the distance $d$ is reduced to a half, in terms of its capacitance, voltage, electric field, and stored energy. Repeat the same analysis when the capacitor is connected to a battery.
    * Capacitor is isolated so $Q$ is constant. $C'=2C$ and $V'=V/2$. $E'=V'/d'=E$ and $U'=1/2QV'=U/2$ but $u'=1/2\epsilon_0E'^2=u$.

## Problem 2
Positive charge Q is uniformly distributed along the positive $y$-axis between as indicated in the figure. Another point charge $q$ is located on the $x$-axis at $x=a$. Calculate the electric field experienced by $q$.
<center><img src="module1ap2.png" width="500" /></center>

*Answer*
One way: calculate the electric potential and then take the gradient. Here we rely on a direct integration of the electric field. Two approaches to get the same answer seen in the workshop, a "geometric approach" and a "systematic approach" (I personally prefer the latter one).

The geometric approach:
The **module** of the electric field at $x$ created by an infinitesimal charge $dq$ located at $y$ is 

$$ d E=\frac{1}{4 \pi \epsilon_{0}}\frac{dq}{x^2+y^2}$$

What is $dq$? As we have a linear charge distribution, then by definition $dq=\lambda dL$. And what is dL in this case? dL=dy. Additionally, it's straightforward to notice that $\lambda=Q/L$ in this case. Thus,

$$ dE=\frac{Q}{4 \pi \epsilon_{0}L}\frac{dy}{x^2+y^2}$$

Now, in looking at the figure, the components of the vector (the electric field is a vector, not its module!) must be

$$
\begin{array}{l}
d E_{x}=d E \cos \theta=\frac{Q x}{4 \pi \epsilon_{0} L}\left(\frac{d y}{\left(x^{2}+y^{2}\right)^{3 / 2}}\right) \\
d E_{y}=-d E \sin \theta=-\frac{Q}{4 \pi \epsilon_{0} L}\left(\frac{y d y}{\left(x^{2}+y^{2}\right)^{3 / 2}}\right) \\
\end{array}
$$

where 

$$
\begin{array}{l}
\sin \theta=\frac{y}{\sqrt{x^{2}+y^{2}}} \\
\cos \theta=\frac{x}{\sqrt{x^{2}+y^{2}}}
\end{array}
$$

Integrating we obtain the electric field ($\vec E=E_x\hat i+E_y \hat j$)

$$
\begin{array}{l}
E_{x}=\int d E_{x}=\frac{Q x}{4 \pi \epsilon_{0} L} \int_{0}^{L} \frac{d y}{\left(x^{2}+y^{2}\right)^{3 / 2}}=\frac{Q x}{4 \pi \epsilon_{0} L}\left[\frac{1}{x^{2}} \frac{y}{\sqrt{x^{2}+y^{2}}}\right]_{0}^{L}=\frac{Q}{4 \pi \epsilon_{0} x} \frac{1}{\sqrt{x^{2}+L^{2}}} \\
E_{y}=\int d E y=-\frac{Q}{4 \pi \epsilon_{0} L} \int_{0}^{L} \frac{y d y}{\left(x^{2}+y^{2}\right)^{3 / 2}}=-\frac{Q}{4 \pi \epsilon_{0} L}\left[-\frac{1}{\sqrt{x^{2}+y^{2}}}\right]_{0}^{L}=-\frac{Q}{4 \pi \epsilon_{0} L}\left(\frac{1}{x}-\frac{1}{\sqrt{x^{2}+L^{2}}}\right)
\end{array}
$$

The systematic approach:

The general expression for the electric field created by $dq$ is

$$ d\vec E=\frac{1}{4 \pi \epsilon_{0}}\frac{dq(\vec r-\vec r')}{|\vec r-\vec r'|^3}$$

Let's identify terms

$$\begin{array}
\vec r=x\hat i \\
\vec r'=y\hat j \\
\vec r-\vec r'=x\hat i-y\hat j \\
|\vec r-\vec r'|^3=(x^2+y^2)^{3/2} \\
dq=\lambda dL=Qdy/L \\
\end{array}$$

Hence

$$ d\vec E=\frac{Q}{4 \pi \epsilon_{0}L}\frac{dy(x\hat i-y\hat j)}{(x^2+y^2)^{3/2}}$$

Integrating the previous expression we obtain the same results as before.

## Problem 3
Calculate the electric field at any point created by an infinitely large cylinder of radius $R$ with a charge density 

$$\begin{array}{ll}\rho(r)=\rho_{0}(1-r / R) & \text { for } r \leq R \\ \rho(r)=0 & \text { for } r \geq R\end{array}$$

where $r$ is the distance from the axis of the cylinder. 

*Answer*:

By choosing a cylinder of radius $r$ as a gaussian surface (over which the electric field is constant), applying the Gauss law we obtain 

$$\int \vec E\cdot dS=E(2\pi rL)=Q_{enc}/\epsilon_0$$
$r>R$:
$$Q_{enc}=Q_{Total}=\int_0^R\rho(r)dV=\rho_0\int_0^R(1-r/R)(2\pi rLdr)=2\pi\rho_0L(\frac{R^2}{2}-\frac{R^3}{3R})=\pi/3\rho_0LR^2$$
$$E(2\pi rL)=\pi\rho_0 LR^2/(3\epsilon_0)\implies \vec E=\frac{\rho_0R^2}{6\epsilon_0r}\hat r$$
$r<R$:
$$Q_{enc}=\int_0^r\rho(r')dV=\rho_0\int_0^r(1-r'/R)(2\pi r'Ldr')=2\pi\rho_0L(\frac{r^2}{2}-\frac{r^3}{3R})$$
$$E(2\pi rL)=\frac{2\pi\rho_0L}{\epsilon_0}(\frac{r^2}{2}-\frac{r^3}{3R})\implies \vec E=\frac{\rho_0}{\epsilon_0}(\frac{r}{2}-\frac{r^2}{3R})\hat r$$

## Problem 6    
    
A parallel-plate capacitor with area $A=500cm^2$ and separation $d=5mm$ has a voltage of $V_0=2000V$. Introducing a dielectric between its plates reduces the voltage to $V=1000V$. Find

* a) The capacity after introducing the dielectric.
* b) The relative permittivity.
* c) The induced charge on each surface of the dielectric.

*Answer*:

* a) Before we introduce the dielectric, we can write $C_0=Q_0/V_0$ and $C_0=\epsilon_0A/d$. From these two equations we can  obtain the charge $Q_0=(\epsilon_0A/d)V_0=1.7\times 10^{-7}C$. 

Once the dielectric is introduced, the charge on the metal plates remains the same ($Q_0$), although the difference of potential does change (bear in mind that the capacitor is isolated,not connected to any battery). Therefore $C=Q_0/V=1.7\times 10^{-7}C/1000V=1.7\times 10^{-4}\mu F$
* b) Applying the definition of relative permittivity, $\epsilon_r=C/C_0=V_0/V=2$.
* c) The dielectric gets polarized after being introduced between the plates of the capacitor. This means that a new superficial charge distribution appears on the surface of the dielectric, $\sigma_i$, with opposite sign with respect to the charge of the metal plate. The module of the electric field between the plates will be now

$$E=\frac{\sigma-\sigma_i}{\epsilon_0},$$

where $\sigma=Q_0/A$ and $\sigma_i=Q_i/A$. 

The relative permittivity can be calculated from 

$$\epsilon_r=\frac{V_0}{V}=\frac{E_0}{E}=\frac{\sigma}{\sigma-\sigma_i}\implies\sigma_i=\sigma\left(1-\frac{1}{\epsilon_r}\right)\implies Q_i=Q_0\left(1-\frac{1}{\epsilon_r}\right)=8.5\times 10^{-7}C$$

## Problem 7

Suppose we connect a couple of capacitors in series, with capacitance of $150nF$ and $120nF$ respectively. Knowing that the potential difference across them is $36V$. Find:

* a) the total charge stored in the network. 
* b) the charge on each capacitor. 
* c) the total energy stored in the network. 
* d) the energy stored in each capacitor. 
* e) the potential differences across each capacitor.

*Answer*

The capacitors are in series, so the overall capacitance of the network is

$$
C_{\mathrm{eq}}=\frac{C_{1} C_{2}}{C_{1}+C_{2}}=\frac{(150 \mathrm{nF})(120 \mathrm{nF})}{150 \mathrm{nF}+120 \mathrm{nF}}=66.7 \mathrm{nF}
$$

a) The total charge is then $$
Q=C V=(66.7 \mathrm{nF})(36 \mathrm{~V})=2.4 \times 10^{-6} \mathrm{C}=2.4 \mu \mathrm{C}
$$

b) As they are in series, each capacitor will store the same amount of electrical charge on its plates regardless of its capacitance. This is because the charge stored by a plate of any one capacitor must have come from the plate of its adjacent capacitor. $Q_1=Q_2=2.4 \mu \mathrm{C}$

c) $$
U=\frac{1}{2} C_{\mathrm{eq}} V^{2}=\frac{1}{2}\left(66.7 \times 10^{-9} \mathrm{~F}\right)(36 \mathrm{~V})^{2}=43.2 \mu \mathrm{J}
$$

d) We know $C$ and $Q$ for each capacitor, and we can write $U$ in term of these

$$
U=\frac{1}{2} C V^{2}=\frac{1}{2} C(Q / C)^{2}=Q^{2} / 2 C
$$

$$
150 \mathrm{nF}: U=\frac{\left(2.4 \times 10^{-6} \mathrm{C}\right)^{2}}{2\left(150 \times 10^{-9} \mathrm{~F}\right)}=19.2 \mu \mathrm{J} ; 120 \mathrm{nF}: U=\frac{\left(2.4 \times 10^{-6} \mathrm{C}\right)^{2}}{2\left(120 \times 10^{-9} \mathrm{~F}\right)}=24.0 \mu \mathrm{J}
$$

The sum of these two energies is obviously the total energy from c)

e) $$
150 \mathrm{nF}: V=\frac{Q}{C}=\frac{2.4 \times 10^{-6} \mathrm{C}}{150 \times 10^{-9} \mathrm{~F}}=16 \mathrm{~V} ; 120 \mathrm{nF}: V=\frac{Q}{C}=\frac{2.4 \times 10^{-6} \mathrm{C}}{120 \times 10^{-9} \mathrm{~F}}=20 \mathrm{~V}
$$

Notice that these values add up to $36V$, the voltage across the capacitors in series.

## Problem 8

Electronic flash units for cameras contain a capacitor for storing the energy used to produce the flash. In one such unit, the
flash lasts for $(1/675)$ seconds with an average light power output of $2.7\times 10^5 W$. 

* a) If the conversion of electrical energy to light is 95% efficient (the rest of the energy goes to thermal energy), how
much energy must be stored in the capacitor for one flash?  
* b) If the capacitor has a potential difference between its plates of 125V, what is the capacitance?

*Answer*

a) The power output is $2.70 \times 10^5 W$ and 95% of the original energy is converted. Thus,

$$
E=P t=\left(2.70 \times 10^{5} \mathrm{~W}\right)\left(1.48 \times 10^{-3} \mathrm{~s}\right)=400 \mathrm{~J} . \quad \implies U=\frac{400 \mathrm{~J}}{0.95}=421 \mathrm{~J}
$$

b) $$
U=\frac{1}{2} C V^{2} \implies C=\frac{2 U}{V^{2}}=\frac{2(421 \mathrm{~J})}{(125 \mathrm{~V})^{2}}=0.054 \mathrm{~F}
$$
