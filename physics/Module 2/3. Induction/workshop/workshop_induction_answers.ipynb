1.Two solenoids with self-inductances $L_1,L_2$ and mutual inductance $M$ are connected in series. What is the net self-inductance of the system?

*Answer*:
Current is $I_1=I_2=I$ and flux is $\Phi=\Phi_1+\Phi_2=L_1I+L_2I+ 2MI.$ Hence. $L=\Phi/I=L_1+L_2+ 2M$.

### Problem 2
A loop moves along the x-axis at a constant speed $v$. A constant current $I$ flown in a long wire, as shown in the figure. Calculate the emf induced in the loop. 

 <center><img src="module2ap2_.png" width="400"/></center>


*Answer*: $B=\mu_0 I/(2\pi r)$. The total flux through the loop is $$
\Phi_{B}=\int d \Phi_{B}=\left(\frac{\mu_{0} I b}{2 \pi}\right) \int_{r}^{r+a} \frac{d x}{x}
$$
$$
\begin{array}{l}
\Phi_{B}=\left(\frac{\mu_{0} I b}{2 \pi}\right) \ln \left(\frac{r+a}{r}\right) \\
\varepsilon=-\frac{d \Phi_{B}}{d t}=-\frac{d \Phi_{B}}{d r} \frac{d r}{d t}=\frac{\mu_{0} I b}{2 \pi}\left(\frac{a}{r(r+a)}\right) v
\end{array}
$$ 
where $$r(t)=r_0+vt$$
We chose the normal vector going into the plane (when calculating the flux), and using the right-hand rule with the normal, as $\varepsilon>0$, the induced current $I=\varepsilon/R$ must be clockwise. Alternatively, $|\varepsilon|=\frac{\mu_{0} I a b v}{2 \pi r(r+a)}$ and using Lenz's law, the current must be clockwise.

### Problem 3
A circular loop of flexible iron wire has an initial circumference of $165 cm$, but its circumference is decreasing at a constant rate of $12cm/s$ due to a tangential pull on the wire. The loop is in a constant, uniform magnetic field oriented perpendicular to the plane of the loop and with magnitude $0.5 T$. Find the emf induced in the loop at the instant when $9s$ have passed.

*Answer*: the circumference is $c=2\pi r$, so the area can be expressed as $A=\pi r^2=c^2/(4\pi)$. Using Faraday's law $$
\begin{array}{l}
\Phi_{B}=B A=(B / 4 \pi) c^{2} \\
|\varepsilon|=\left|\frac{d \Phi_{B}}{d t}\right|=\left(\frac{B}{2 \pi}\right) c\left|\frac{d c}{d t}\right|
\end{array}
$$
$$
\begin{array}{l}
\text { At } t=9.0 \mathrm{~s}, c=1.650 \mathrm{~m}-(9.0 \mathrm{~s})(0.120 \mathrm{~m} / \mathrm{s})=0.570 \mathrm{~m} \\
|\varepsilon|=(0.500 \mathrm{~T})(1 / 2 \pi)(0.570 \mathrm{~m})(0.120 \mathrm{~m} / \mathrm{s})=5.44 \mathrm{mV}
\end{array}
$$

Using Lenz's law, the current is clockwise. Alternatively, notice that $\varepsilon>0$, so using the right-hand rule with the normal vector going into the plane, it must be clockwise.

### Problem 4

A long, thin solenoid has $900$ turns per meter and radius $2.50 cm$. The current in the solenoid is increasing at a uniform rate of $60A/s$. What is the magnitude of the induced electric field at $0.5 cm$ from the axis of the solenoid?


*Answer*: Let's apply Faraday's law to a circle of radius $r$. $$
\oint \vec{E} \cdot d \vec{l}=-\frac{d \Phi_{B}}{d t}
$$
$$
\begin{array}{l}
\Phi_{B}=B \pi r^{2} \cdot\left|-\frac{d \Phi_{B}}{d t}\right|=\pi r^{2}\left|\frac{d B}{d t}\right| \\
|\oint \vec{E} \cdot d \vec{l}|=\left|-\frac{d \Phi_{B}}{d t}\right| \implies E(2 \pi r)=\pi r^{2}\left|\frac{d B}{d t}\right| \\
E=\frac{1}{2} r\left|\frac{d B}{d t}\right| \\
B=\mu_{0} n I, \mathrm{so} \frac{d B}{d t}=\mu_{0} n \frac{d I}{d t} \\
\implies E=\frac{1}{2} r \mu_{0} n \frac{d I}{d t}=\frac{1}{2}(0.00500 \mathrm{~m})\left(4 \pi \times 10^{-7} \mathrm{~T} \cdot \mathrm{m} / \mathrm{A}\right)\left(900 \mathrm{~m}^{-1}\right)(60.0 \mathrm{~A} / \mathrm{s})\\
=1.70 \times 10^{-4} \mathrm{~V} / \mathrm{m} .
\end{array}
$$
