# 1. p-n junction: structure

What is it? A p-n junction consists of a union a n-type and p-type semiconductors (a single semiconductor crystal with two doped regions). 

<center><img src="https://upload.wikimedia.org/wikipedia/commons/thumb/7/79/PN_diode_with_electrical_symbol.svg/1920px-PN_diode_with_electrical_symbol.svg.png" width="400" /></center>

*Basic properties*:

* Depending on the voltage applied, the junction between the two doped regions can become depleted of mobile charge carriers (called **depletion region**), leaving ionized impurities left, hence becoming non-conductive.
* As a result of having a depletion region, a p–n junction diode allows electric charges to flow in one direction, but not in the opposite direction; electrons can easily flow through the junction from n to p but not from p to n, while the reverse holds for holes. A **diode** is a circuit element defined by **allowing a flow of electricity in one direction but not the the opposite**. **Forward bias** is defined in the direction of easy current flow, while **reverse bias** is in the direction of little or no current flow.

The basic mathematical model of a p-n junction was developed by Shockley in 1949 [link](https://onlinelibrary.wiley.com/doi/abs/10.1002/j.1538-7305.1949.tb03645.x).

It's crucial to understand how a p-n junction works, because it is the essential building block used in practically all kind of electronic devices: diodes, transistors, solar cells, LEDs, rectifiers...

# 2. p-n junction in equilibrium

## 2.1 Reaching the steady state

I. Before the junction, we have two separate band diagrams, with two different Fermi levels. We assume no degeneracy and no electric field.

<center><img src="eq1.png" width="500" /></center>

|  n-type    |  p-type    |
|------|------|
|$$ \label{eqN}n_{n_0}=N_{c} e^{\frac{E_{F}-E_{c}}{k_{B} T}} \\p_{n_0}=n_{i}^{2} / n_{n_0}$$|$$ \label{eqP}p_{p_0}=N_{v} e^{\frac{E_{v}-E_{F}}{k_{B} T}} \\n_{p_0}=n_{i}^{2} / p_{p_0}$$|

We shall define $n_{n_0},p_{n_0}$ and $n_{p_0},p_{p_0}$ as the values of the charge carrier densities, for the n and p side of the junction respectively, far away from the junction.

II. After the junction, there exists a strong **gradient of concentration** of impurities, which originates a current flux of carriers
* electrons $e^-$ from n to p
* holes $h^+$ from p to n

III. Such diffusion current holds as long as there exists a gradient of concentrations. Each $e^-$ that abandons the n-type region leaves behind a fixed positive ion. Similarly, each $h^+$ leaves behind a negative ion. As a result, a **depletion region** is created around the junction where there is a spatial separation of positive and negative charges, with **charge density** $\rho(x)$.

<center><img src="eq2.png" width="500" /></center>

IV. Such charge density $\rho(x)$ leads to an **electric field** $\vec E$ from n to p. Such electric field will **drift** electrons $e^-$ from p to n and holes $h^+$ from n to p. 
<center><img src="eq3.png" width="500" /></center>

<center><img src="https://upload.wikimedia.org/wikipedia/commons/thumb/3/3e/Pn_Junction_Diffusion_and_Drift.svg/800px-Pn_Junction_Diffusion_and_Drift.svg.png" width="500" /></center>

IV. At equilibrium, we reach a ***dynamical balance*** by which diffusion currents equal drift currents, 

$$\left.J_{n}\right|_{\text{dri}}=\left.J_{n}\right|_{\text{dif}}\\
\left.J_{p}\right|_{\text{dri}}=\left.J_{p}\right|_{\text{dif}}$$

with a constant depletion region around the junction of width $W$. Such **inhomogeneous charge distribution** is associated with an electric field, and consequently, a potential difference across the junction, called **built-in potential** $V_{bi}$.

<center><img src="eq4.png" width="400" /></center>

Convention: we define $V(x)=0$ on the n side: all potential is applied to the p side, so $V_{bi}=V_p-V_n>0$

## 2.2 Band diagram in equilibrium

Recall: a band diagram is a representation of the energy of electrons.

<center><img src="bands.png" width="400" /></center>

The Fermi level must be constant along the junction. There must be a difference of energy between the two regions due to the built-in potential, which makes energy of $e^-$ in the n-type semiconductor diminish with respect to electrons in the p-type (recall potential and energy have opposite signs!). 

We define $E_{C_n}=0$ far away from the junction on the n-type side, and therefore $E_{C_p}=-q V_{bi}=eV_{bi}>0$ on the p-type side.

## 2.3 Built-in potential

Far away from the junction, the carrier concentrations will be given by equations from 2.1 . In looking at the energy band diagram, we readily notice that


$$
\begin{aligned}
-e V_{b i} &=-e(V_{p}-V_{n})=E_{C_n}-E_{C_p}=-k_{B} T \ln \left(\frac{n_{n_0}}{n_{p_0}}\right)=-k_{B} T \ln \left(\frac{p_{p_0}}{p_{n_0}}\right)
\end{aligned}
$$

As the mass-action law holds ($n_{p_0}p_{p_0}=n_{n_0}p_{n_0}=n_i^2$), and assuming all impurities are ionized, we conclude that

|  |
|--|
|$$V_{b i} =\frac{k_{B} T}{e} \ln \left(\frac{n_{n_0}p_{p_0}}{n_i^2}\right)\simeq\frac{k_{B} T}{e} \ln \left(\frac{N_{A} N_{D}}{n_i^2}\right)$$|
|   |

Since $V_{b i}>0\implies N_D,N_A>n_i$. 

Also notice that $V_{b i}<E_g$

Typically, at room temperature the voltage across the depletion layer for silicon is about 0.6 – 0.7 volts and for germanium is about 0.3 – 0.35 volts.

$V_{th}=\frac{k_{B} T}{e}$ is usually called *thermal voltage*, which is 26mV at room temperature

## 2.4. Regions of the junction

### 2.4.1 neutral region

Far away from the junction, we find neutrality charge at the microscopic and macroscopic level, so **there is no electric field**. Charge carrier concentrations, $n_{n0},n_{p0},p_{n0},p_{p0}$, are given by equations in 2.1. All impurities are ionized.

### 2.4.2 depletion region

Macroscopically neutral, but **not** microscopically: charge carrier density is a function of position across the junction, $n(x),p(x)$. Namely, $n(x)$ goes from $N_D$ in they n-type region (majority carriers) to $n(x)=n_i^2/n_A$ in the p-type region (minority carriers). Analogously for $p(x)$.

One can calculate (see appendix for proof) the limits of the depletion region:

$$
x_{p}=\sqrt{\frac{2 \varepsilon V_{b i}}{e N_{A}\left(1+\frac{N_{A}}{N_{D}}\right)}}
$$
$$
x_{n}=\sqrt{\frac{2 \varepsilon V_{b i}}{e N_{D}\left(1+\frac{N_{D}}{N_{A}}\right)}}
$$

A **one-sided p-n junction** is defined as the junction in which one region is highly doped compared to the other one (denoted $n^+p$ or $p^+n$). In this junctions, the depletion region lies almost in the less doped region.

And finally, an expression for the **depletion width** is obtained

|   |
|---|
|$$W=\sqrt{\frac{2 \varepsilon V_{b i}}{e} \frac{N_{A}+N_{D}}{N_{A} N_{D}}} $$|

The width of depletion region is of the order of μm

# 3. Polarized p-n junction

What are the effects of applying an electric field $E$ on a p-n junction? Ideally, the study can be separated into independent effects,

* The effect on the depletion region.
* The effect on the neutral region.

In reality, currents calculated in the neutral region are affected by the depletion region (considered as deviations from ideal behavior).

We continue assuming that the electric field $E$ implied by the externally applied voltage lies in the depletion region, so **the electric field in the neutral region is zero**.

## 3.1 Band diagram

**Forward bias**: applying a positive voltage V (to the p-type side). Net voltage will be $=V-V_{bi}$.
<center><img src="polarized2.png" width="400" /></center>

We decrease the net potential in the depletion region $\implies$ net $E$ decreases $\implies$ width $W$ decreases. 

Equilibrium was reached when the built-in electric field compensated diffusion. The external electric field opposes this field, **favoring diffusion** $\implies$ net current $I\neq0$; current flows **from p to n**

Quasi Fermi levels
$F_n>F_p$. Also $V=F_n-F_p$

<center><img src="polarized1.png" width="700" /></center>



**Reverse bias**: applying a negative voltage (p region). Net voltage$=V-(-V_{bi})$.
<center><img src="polarized4.png" width="400" /></center>

Opposite case 
* Quasi Fermi levels $F_p>F_n$
* $W$ increases
* applied electric field opposes diffusion. $I=0$

<center><img src="polarized3.png" width="700" /></center>


## 3.2 Effects in depletion region

### 3.2.1 depletion width

The previous equations hold, except we replace the net potential


$$
x_{p}=\sqrt{\frac{2 \varepsilon (V_{b i}-V)}{e N_{A}\left(1+\frac{N_{A}}{N_{D}}\right)}}
$$
$$
x_{n}=\sqrt{\frac{2 \varepsilon (V_{b i}-V)}{e N_{D}\left(1+\frac{N_{D}}{N_{A}}\right)}}
$$


|   |
|---|
|$$W=\sqrt{\frac{2 \varepsilon (V_{b i}-V)}{e} \frac{N_{A}+N_{D}}{N_{A} N_{D}}}$$|
|   |

* $V>0\implies W$ decreases
* $V<0\implies W$ increases

when $V>V_{bi}$, the depletion width vanishes. 

### 3.2.2 depletion capacitance

The junction has also associated a capacitance

|   |   |   |
|---|---|---|
|<img src="capacitance1.png" width="500" />|<img src="capacitance2.png" width="500" />|<img src="capacitance3.png" width="500" />|

Due to charge neutrality we have,
$$
|Q|=|e  N_{A} x_{p}|=|e{N}_D x_{n}|
$$

Then,
$$
C_{T}=\frac{d |Q| }{d V}=e N_{A} \frac{d x_p}{d V}=e N_{D} \frac{d x_{n}}{d V}=\sqrt{\frac{e \varepsilon}{2} \frac{N_A N_{D}}{N_A+N D}} \frac{1}{\sqrt{V_{bi}-V}}=\frac{\varepsilon}{W}
$$

same as a plane capacitor of width $W$. 


$C_{dep}=\frac{A\varepsilon}{W}$

# 4. Characteristic curve I-V

## 4.1 Assumptions Shockley model
As result of polarizing the diode, $V\neq0$, there appears a current. We shall calculate this current using the Shockley model, which assumes:

* unidimensional junction: $\nabla=\frac{\partial}{\partial x}$.
* full-depletion approximation $\implies$ all electric field lies in depletion region.
* Boltzmann approximation (no degenerate semiconductors)
* long diode: far away from junction we recover equilibrium conditions.
* low injection regime $\implies$ $p_n\ll n_{n_0}$ in n region and $n_p\ll p_{p_0}$ in p region 
* no generation or recombination within the depletion region ($W\ll L_p,L_n$), which implies
    - net current is constant along the junction
    - constant currents of $e^-$ and $h^+$ in the depletion region.
   

<center><img src="currents.png" width="600" /></center>

## 4.2 boundary conditions for minority carriers

In order to calculate the density carrier concentration along the junction, we shall need to impose certain boundary conditions, at infinity and between the neutral and depletion regions.

$$n p=N_{C} \exp \left(\frac{F_{n}-E_{c}}{k_{B} T}\right) N_{V} \exp \left(\frac{E_{v}-F_{p}}{k_{B} T}\right)=\\=n_{i}^{2} \exp \left(\frac{F_{n}-F_{p}}{k_{B} T}\right)=n_{i}^{2} \exp \left(\frac{eV}{k_{B} T}\right)$$

Then

$$p\left(x_{n}\right)=\frac{n_{i}^{2}}{N_{D}} \exp \left(\frac{e V}{k_{B} T}\right)=p_{n_0} \exp \left(\frac{e V}{k_{B} T}\right)$$

$$
n\left(-x_{p}\right)=\frac{n_{i}^{2}}{N_{A}} \exp \left(\frac{e V}{k_{B} T}\right)=n_{p_0} \exp \left(\frac{e V}{k_{B} T}\right)
$$

## 4.3 minority carriers profile

We are going to calculate the minority carrier density using the drift-diffusion equation. Mobile carriers were neglected when calculated the electric field in the depletion region, but cannot be neglected when calculating the currents.

**n-type region**

$$
D_{p} \frac{\partial^{2} p_{n}}{\partial x^{2}}-\mu_{p} E \frac{\partial p_{n}}{\partial x}-\mu_{p} p_{n} \frac{\partial E}{\partial x}+G-R=\frac{\partial p_{n}}{\partial t}
$$
* $E=0 ; \frac{\partial E}{\partial x}=0$
* $G=0$ (only $G_{th}=R_{th}$)
* $R=\frac{p_{n}-p_{n 0}}{\tau_{p}}$ (low injection)
* $\frac{\partial p_{n}}{\partial t}=0$ (steady state)
* boundary conditions
    $$p_n(\infty)=p_{n0}$$
    $$p_n(x_n)=p_{n0}\exp(\frac{eV}{kT})$$

The equations comes down to
\begin{equation}
\frac{\partial^{2} p_{n}}{\partial x^{2}}-\frac{p_{n}-p_{n 0}}{L_{p}^{2}}=0
\end{equation}

\begin{equation}
L_{p}=\sqrt{D_{p} \tau_{p}}
\end{equation}

diffusion length (the mean distance a carrier diffuses before recombination) $L_{p}$ is typically on the order of micrometers.

Solution
\begin{equation}
p_{n}(x)=p_{n_0} \cdot\left[\exp \left(\frac{e V}{k_{B} T}\right)-1\right] \exp \left(\frac{x_{n}-x}{L_{p}}\right)+p_{n_0}
\end{equation}

Similarly for **p-type region**

* boundary conditions
$$n_p(-\infty)=n_{p0}$$
$$p_n(-x_p)=n_{p0}\exp(\frac{eV}{kT})$$

Solution:
\begin{equation}
n_{p}(x)=n_{p_0}\left[\exp \left(\frac{e V}{k_{B} T}\right)-1\right] \exp \left(\frac{x_{p}+x}{L_{n}}\right)+n_{p_0}
\end{equation}

<center><img src="profile.png" width="500" /></center>

## 4.4 Shockley ideal diode equation

As we have assumed no electric field in the neutral region, in the **neutral region** there will be **only a diffusion current of minority carriers**.

n-type region:

\begin{equation}
J_{p} \cong \left.J_{}\right|_{\text{dif}}=-e D_{p} \frac{\partial p_{n}}{\partial x}=\frac{e D_{p} p_{n 0}}{L_{p}} \cdot\left[\exp \left(\frac{e V}{k_{B} T}\right)-1\right] \exp \left(\frac{x_{n}-x}{L_{p}}\right)
\end{equation}

p-type region:
\begin{equation}
J_{n}=\frac{e D_{n} n_{p 0}}{L_{n}} \cdot\left[\exp \left(\frac{e V}{k_{B} T}\right)-1\right] \exp \left(\frac{x_{p}+x}{L_{n}}\right)
\end{equation}

Net current

\begin{equation}
J=\left.J_{p}\right|_{x=x_{n}}+\left.J_{n}\right|_{x=-x_{p}}=J_{s}\left[\exp \left(\frac{e V}{k_{B} T}\right)-1\right]
\end{equation}

reverse bias saturation density current

$$
J_{s}=e n_{i}^{2}\left[\frac{D_{p}}{N_{D} L_{p}}+\frac{D_{n}}{N_{A} L_{n}}\right]=\frac{e D_{p} p_{n 0}}{L_{n}}+\frac{e D_{n} n_{p 0}}{L_{n}}
$$

$I_s=J_s A$ maximum current with reverse bias

Shockley diode equation (characteristic curve I-V)

|  |
|--|
|$$I=JA=I_{s}\left[\exp \left(\frac{e V}{k_{B} T}\right)-1\right]$$|
|   |

the voltage and the current do not have a linear relationship, nonohmic device!

## 6.3 How to solve circuits with diodes?

We shall assume the piecewise-linear approximation, by which we will be able to leverage the methods learned for linear circuits (like mesh analysis) when the diode is assume to be functioning in a given regime.

1. Make an hypothesis about the functioning of the diode.
2. Replace the diode by the corresponding circuit model (voltage source or open circuit).
3. Solve the circuit.
4. Check whether the result is consistent with the condition for the functioning selected.
5. In case of not being consistent, restart the process with another hypothesis.

## Electric field and potential along the junction

General expression for charge density will be given by charge neutrality condition $
\rho(x)=e\left[-n+N_{D}^{+}+p-N_{A}^{-}\right]
$. Here we shall employ the **full-depletion approximation**, by which we neglect mobile carriers and the charge density is given by

$$
\rho(x) \simeq\left\{\begin{array}{cc}
e N_{D} & -x_{n} \leq x \leq 0 \\
-e N_{A} & 0 \leq x \leq x_{p} \\
0 & \text { rest }
\end{array}\right.
$$

<img src="eq2.png" width="300" />

Our goal is to calculate $E(x),V(x),x_n,x_p$. This can be down thanks to Poisson's equation.

$$
-\frac{\partial^{2} \phi}{\partial x^{2}}=\frac{\partial E}{\partial x}=\frac{\rho(x)}{\varepsilon}
$$

Integrating once we obtain the electric field

$$
E(x)=\left\{\begin{array}{c}
\frac{e N_{D}}{\varepsilon}\left(x+x_{n}\right)-x_{n} \leq x \leq 0 \\
\frac{e N_{A}}{\varepsilon}\left(-x+x_{p}\right) \quad 0 \leq x \leq x_{p} \\
0 \quad \text { rest }
\end{array}\right.
$$

<img src="eq3.png" width="300" />

Continuity of electric field at $x=0$ (no free charge!) implies $N_{D} x_{n}=N_{A} x_{p}$

Integrating twice we obtain the potential (imposing the boundary condition $V(-\infty)=0$)

$$
V(x)=\left\{\begin{aligned}
-\frac{e N_{D}\left(x_{n}+x\right)^{2}}{2 \varepsilon}-x_{n} \leq x & \leq 0 \\
-\frac{e N_{D} x_{n}^{2}}{2 \varepsilon}-\frac{e N_{A}\left(2 x x_{p}-x^{2}\right)}{2 \varepsilon} & 0 \leq x \leq x_{p}
\end{aligned}\right.
$$

Particularizing this expression at $x_p$, we obtain the built-in potential

|   |
|---|
|$$
V_{b i}=\frac{e}{2 \varepsilon}\left(N_{D} x_{n}^{2}+N_{A} x_{p}^{2}\right)
$$|


<img src="pot.png" width="300" />

As $x_{n} N_{D}=x_{p} N_{A}$

$$
x_{p}=\sqrt{\frac{2 \varepsilon V_{b i}}{e N_{A}\left(1+\frac{N_{A}}{N_{D}}\right)}}
$$
$$
x_{n}=\sqrt{\frac{2 \varepsilon V_{b i}}{e N_{D}\left(1+\frac{N_{D}}{N_{A}}\right)}}
$$

A **one-sided p-n junction** is defined as the junction in which one region is highly doped compared to the other one (denoted $n^+p$ or $p^+n$). In this junctions, the depletion region lies almost in the less doped region.

And finally, an expression for the depletion width is obtained

|   |
|---|
|$$W=\sqrt{\frac{2 \varepsilon V_{b i}}{e} \frac{N_{A}+N_{D}}{N_{A} N_{D}}}$$|

## Currents and quasi Fermi levels

We can express the electric field as

$$
\vec{E}=-\nabla V=-\nabla\left(-\frac{E_{c}}{q}\right)=\frac{1}{e} \frac{\partial E_{c}}{\partial x}
$$

Bearing in mind $n=N_{c} e^{\frac{F_{n}-E_{c}}{k_{B} T}}$, then

\begin{equation}
J_{n}=e \mu_{n}\left(n E+\frac{k_{B} T}{e} \frac{\partial n}{\partial x}\right)=e \mu_{n}\left(N_{c} e^{\frac{F_{n}-E_{c}}{k_{B} T}} \frac{1}{e} \frac{\partial E_{c}}{\partial x}+\frac{N_{c}}{e} e^{\frac{F_{n}-E_{c}}{k_{B} T}} \frac{\partial F_{n}}{\partial x}-\frac{N_{c}}{e} e^{\frac{F_{n}-E_{c}}{k_{s} T}} \frac{\partial E_{c}}{\partial x}\right)=\mu_{n} n \frac{\partial F_{n}}{\partial x}
\end{equation}

Analogously $J_{p}=\ldots=\mu_{p} p \frac{\partial F_{p}}{\partial x}$. So the net current is

$J=\mu_{n} n \frac{\partial F_{n}}{\partial x}+\mu_{p} p \frac{\partial F_{p}}{\partial x}$

Conclusions:

* In the absence of electric field, there is no current so Fermi levels are constant (as we knew...). 
* If we assume constant currents within the depletion region, then the gradients of Quasi Fermi levels are constant.
