```python
# calculating real part of circuit impedance in ohms
import numpy as np
import math

# Defining omega (w)
omega = 2 * np.pi * 50
micro = 1*10**-6
mili = 1*10**-3
# one resistor and one inductor in parallel
zResistor = (1 / 4)
zInductor1 = omega * (mili*30)
zParallel = complex(zResistor, zInductor1)
print(f"The impedance of the parallel section is {(1 / zParallel):.2} Ohms")
```

```python
zConductor = 1 / (omega * (micro*900))
zInductor2 = omega * (mili*50)
totalImpedance = zInductor2 + zConductor + (1 / zParallel)
print(f"The total impedance is {totalImpedance:.2f}")
```

```python
# Formula for calculating 
# peak current = peak voltage / magnitude of total impedance

# First, use a function to have magnitude and direction in degrees for
# the complex number
def polar(complexNumber):
  # z is a complex number
  # returns a tuple of magnitude and angle in degrees
  mag = abs(complexNumber) # magnitude is the absolute value of z
  ang = math.degrees(math.atan2(complexNumber.imag, complexNumber.real)) # angle is the arctangent of imaginary and real parts in degrees
  return (mag, ang)

# Calculate peak current with the formula
peakVoltage = 220
peakCurrent = peakVoltage / polar(totalImpedance)[0]

# Print the result
print(f"""The peak current is {peakCurrent:.2f} Amps
The phase shift is {polar(totalImpedance)[1]:.2f}""")
```
