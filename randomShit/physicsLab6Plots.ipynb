```python
import numpy as np
import matplotlib.pyplot as plt

# Define the three frequencies
period = 10
f1 = 2 * 3.141516 / (period * 100)
f2 = f1 / 2
f3 = f1 / 6

# Define the x-axis values
x = np.linspace(0, 1000, 1000)

# Calculate the y-axis values for each frequency
y1 = np.cos(2*np.pi*f1*x)
y2 = np.cos(2*np.pi*f2*x)
y3 = np.cos(2*np.pi*f3*x)

# Plot the three cosine waves with different colors
plt.plot(x, y1, 'r', label='Frequency 1')
plt.plot(x, y2, 'g', label='Frequency 2')
plt.plot(x, y3, 'b', label='Frequency 3')

# Add a legend and title
plt.legend()
plt.title('Cosine Waves for Three Frequencies')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the constants
a = 0.017
b = 0.004

# Define the x-axis values
x = np.linspace(2, 400, 1000)

# Calculate the y-axis values for each distance
y = 1 / (a + b * x)

# Plot the frequency as a function of distance
plt.plot(x, y)

# Add labels and title
plt.xlabel('Distance (cm)')
plt.ylabel('Frequency (Hz)')
plt.title('Frequency Generated by Buzzer as a Function of Distance')

# Show the plot
plt.show()
```
